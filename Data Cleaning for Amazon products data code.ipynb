```python
import pandas as pd

# 1. Load the raw Amazon dataset
df = pd.read_csv('amazon_product.csv')
print(df.info())
```

```python
df
```

```python
# 2. Drop unnecessary and mostly empty columns
# 'errors="ignore"' ensures the code doesn't crash if a column is already gone
columns_to_drop = ['Unnamed: 0', 'product_availability', 'unit_price', 'unit_count']
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

```python
# 3. Remove any duplicate product rows
df.drop_duplicates(inplace=True)
```

```python
def clean_price(col_name):
    # Step 1: Grab the specific column from the dataset
    raw_column = df[col_name]

    # Step 2: Force the data to be treated as text (strings) so we can edit the characters
    text_data = raw_column.astype(str)

    # Step 3: Find every '$' symbol and replace it with nothing (deleting it)
    no_dollars = text_data.str.replace('$', '')

    # Step 4: Find every ',' symbol and replace it with nothing (turns 1,000 into 1000)
    no_commas = no_dollars.str.replace(',', '')

    # Step 5: Convert the clean text back into a decimal number (a 'float')
    final_numbers = no_commas.astype(float)

    return final_numbers
```

```python
# Apply the cleaning function to the price columns
df['product_price'] = clean_price('product_price')
df['product_minimum_offer_price'] = clean_price('product_minimum_offer_price')
```

```python
# Step 1: Treat the original price column as text so we can edit the symbols
text_prices = df['product_original_price'].astype(str)

# Step 2: Delete the dollar signs and commas
no_dollars = text_prices.str.replace('$', '')
no_commas = no_dollars.str.replace(',', '')

# Step 3: Convert the text into numbers.
# 'errors="coerce"' means: "If you find text you can't turn into a number, just make it blank (NaN)."
numeric_prices = pd.to_numeric(no_commas, errors='coerce')

# Step 4: Find any blanks (NaN) and fill them with the current price of that exact product
final_original_prices = numeric_prices.fillna(df['product_price'])

# Step 5: Save this fully cleaned column back into our main dataset
df['product_original_price'] = final_original_prices
```

```python
# 6. Handle missing star ratings
# Fill missing ratings with the mathematical average of all other ratings
mean_rating = df['product_star_rating'].mean()
df['product_star_rating'] = df['product_star_rating'].fillna(mean_rating)
```

```python
# 7. Fill missing categorical text data with 'Unknown'
df['sales_volume'] = df['sales_volume'].fillna('Unknown')
df['delivery'] = df['delivery'].fillna('Unknown')
```

```python
# 8. Export the final, cleaned dataset
df.to_csv('cleaned_amazon_product.csv', index=False)
```

```python
# Display a summary to prove the data is clean
print("Data Cleaning Complete!")
print(df.info())
```
